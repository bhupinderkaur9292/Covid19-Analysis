```python
import pandas as pd
import matplotlib.pyplot as plt

# Find top 5 countries by total cases
top5_cases = df.nlargest(5, 'Total Cases')

# Plot
plt.figure(figsize=(8,5))
plt.bar(top5_cases['Country'], top5_cases['Total Cases'], color='skyblue')
plt.title('Top 5 Countries with Highest COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
least5_Cases=df.nsmallest(5,"Total Cases")
least5_Cases
```

```python
plt.figure(figsize=(8,5))
plt.bar(least5_Cases['Country'], least5_Cases['Total Cases'], color='lightcoral')
plt.title('Bottom 5 Countries with Lowest COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
least5_Deaths=df.nsmallest(5,"Total Deaths")
```

```python
plt.figure(figsize=(8,5))
plt.bar(least5_Deaths['Country'], least5_Deaths['Total Deaths'], color='salmon')
plt.title('Bottom 5 Countries with Lowest COVID-19 Deaths')
plt.xlabel('Country')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate death percentage
df['Death Percentage'] = (df['Total Deaths'] / df['Total Cases']) * 100

top5_death_percent = df.nlargest(5, 'Death Percentage')
top5_death_percent
```

```python
plt.figure(figsize=(8,5))
plt.bar(top5_death_percent['Country'], top5_death_percent['Death Percentage'], color='purple')
plt.title('Top 5 Countries by COVID-19 Death Percentage')
plt.xlabel('Country')
plt.ylabel('Death Percentage (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
Continent_Cases=df.groupby("Continent")["Total Cases"].sum().reset_index()
Continent_Cases
```

```python
plt.figure(figsize=(8,5))
plt.bar(Continent_Cases['Continent'], Continent_Cases['Total Cases'], color='teal')
plt.title('COVID-19 Cases by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Population'], df['Total Cases'], color='orange')
plt.title('Relationship between Population and Total COVID-19 Cases')
plt.xlabel('Population')
plt.ylabel('Total Cases')
plt.show()
```
